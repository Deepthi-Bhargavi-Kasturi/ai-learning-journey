```python
#NLP Task: sentiment-analysis
#Output fields: label, score
#Predicted Label: NEGATIVE
#confidence score: 0.989656

from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been trying to avoid last minute plans my whole life.")
```

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)

classifier(
    "This Anime is about Adventure and not giving up",
    candidate_labels=["Motivation","Anime","vacation"]
)

#Multiple sentences - passed as an Array
classifier(
    ["This Show is about Adventure and not giving up","This is a course about the Transformers library"],
    candidate_labels=["Inspiration","Anime","vacation"]
)

##POINTS:
#From the output we can see that, it selected model by default. WHICH IS NOT RECOMMENDED IN PRODUCTION
#When I called classifier object multiple times with different sentences, on execution, last one is picked.
#And to perform classification on multiple sentences -> need to pass sentences in an Array like above
```

```python
#NLP Task: text-generation
#Input: Prompt
#Output: Dictionary with key : generated_text, value: model generated text

from transformers import pipeline

generate = pipeline('text-generation')

generate('In this course, we will teach you how to')
```

```python
#arguments: num_return_sequences and max_length
#num_return_sequences = to control number of different sequences that are generated
#max_length total maximum length of the sequence => (input prompt tokens + newly generated tokens)

from transformers import pipeline

generate = pipeline('text-generation')

generate("Cats are cute creatures. They love their humans. Their favorite food is",
         num_return_sequences=2,
         max_length=10
         )
```

<h1>fill-mask NLP Task</h1>

<u>NLP Task</u>: fill-mask</br>


---


<u>Input</u><br>
**mask token:** is like a place holder in a sentence. Model evaluates and fills it in with the most possible words. </br>
**top_k**: number of words depends on top_k argument value.


---


<u>Output:</u><br>
**score:** confidence score for the predicted word depending on the context of the sentence.<br>
**token:** token id is the **unique** representation of the predicted word in the AI's vocab list.<br>
**token_str:** predicted word

**Ranked possibilities**: The array output is automatically sorted in descending order that is highest score to lowest score. That is ***index = 0 is the model's absolute best guess.***


---

Model used: https://huggingface.co/google-bert/bert-base-uncased$0


```python
#model = bert-base-uncased
#mask token = [MASK]
#top_k = 5 possibilities to be displayed for [MASK]

from transformers import pipeline

unmasker = pipeline('fill-mask', model="bert-base-uncased")
unmasker("Books are a way to escap [MASK]. Any better way to explore the univers!", top_k=5)
```

```python
from transformers import pipeline

#ner = pipeline('ner', grouped_entities=True)
ner = pipeline('ner', aggregation_strategy="simple")
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")

ner2 = pipeline('ner', aggregation_strategy="simple")
ner2("My name is Deepthi Kasturi and I work at Hugging Face and also Google in Brooklyn.")
```

```python
#when we don't pass aggregation_strategy="simple"
#this is equivalent to aggregation_strategy="none"
#BIO Tagging system - how model identifies the entitiy
from transformers import pipeline

ner = pipeline('ner')
ner('My name is Sylvain and I work at Hugging Face in Brooklyn.')
```

```python
#POS - part-of-speech
from transformers import pipeline
pos = pipeline('token-classification', model="vblagoje/bert-english-uncased-finetuned-pos", aggregation_strategy="simple")
pos("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```
