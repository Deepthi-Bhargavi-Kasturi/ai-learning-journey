```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
from transformers import pipeline

#question_answerer = pipeline("question-answering")
question_answerer  =pipeline("text-generation", model="gpt2")
my_context = "The Eiffel Tower is located in Paris, France. It was constructed from 1887 to 1889."
my_question = "When was the Eiffel Tower built?"

# 3. Combine them into a structured prompt string
prompt = f"""Use the following context to answer the question.

Context: {my_context}

Question: {my_question}
Answer:"""
question_answerer(
    prompt,max_new_tokens=15
)
```
