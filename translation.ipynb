```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("Ce cours est produit par Hugging Face.")
```

```python
#english to japanese
from transformers import pipeline

translator = pipeline("translation", model="japanese-denim/mbart-50-finetuned-naga-to-eng")
translator("This course is Hugging Face.")
```

```python
from transformers import pipeline

translator = pipeline("translation_en_to_jp", model="japanese-denim/mbart-50-finetuned-naga-to-eng")
translator("This course is Hugging Face.")


## apparently the model used is wrong It is for Naga to English.
```

```python

from transformers import pipeline

translator = pipeline("translation_en_to_jap", model="Helsinki-NLP/opus-mt-en-jap", max_new_tokens=10)
#translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-jap") -- also works
translator("This course is Hugging Face.")
```
